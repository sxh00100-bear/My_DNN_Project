```python
import tensorflow as tf
from tensorflow import keras

print("=== з¬¬ 1 жӯҘпјҡжЁЎжӢҹдёҖдёӘеҲ«дәәи®ӯз»ғеҘҪзҡ„жЁЎеһӢ (Model A) ===")
# дҝ®жӯЈ 1пјҡз”Ёдё“й—Ёзҡ„ Input еұӮд»ЈжӣҝеҺҹжқҘзҡ„ input_shapeпјҢж¶ҲзҒӯиӯҰе‘ҠпјҒ
model_A = keras.models.Sequential([
    keras.Input(shape=(28, 28)),  # <--- ж–°зүҲжң¬зҡ„ж ҮеҮҶеҶҷжі•
    keras.layers.Flatten(),
    keras.layers.Dense(100, activation="elu", kernel_initializer="he_normal"),
    keras.layers.Dense(50, activation="elu", kernel_initializer="he_normal"),
    keras.layers.Dense(10, activation="softmax") 
])

print("\n=== з¬¬ 2 жӯҘпјҡејҖе§ӢдҪ зҡ„'зҷҪе«–'жүӢжңҜ ===")
model_A_clone = keras.models.clone_model(model_A)
model_A_clone.set_weights(model_A.get_weights())

model_B = keras.models.Sequential(model_A_clone.layers[:-1])
model_B.add(keras.layers.Dense(1, activation="sigmoid"))

print("\n=== з¬¬ 3 жӯҘпјҡи§ҒиҜҒж ёеҝғйӯ”жі•пјҲеҶ»з»“пјү ===")
for layer in model_B.layers[:-1]:
    layer.trainable = False  

# дҝ®жӯЈ 2пјҡжүӢеҠЁе‘ҠиҜү model_B иҫ“е…Ҙж•°жҚ®зҡ„еҪўзҠ¶пјҢејәиҝ«е®ғз«ӢеҲ»жҠҠвҖңи„‘иўӢвҖқз»„иЈ…еҘҪпјҒ
model_B.build(input_shape=(None, 28, 28)) 

model_B.compile(loss="binary_crossentropy", 
                optimizer=keras.optimizers.SGD(learning_rate=0.01), 
                metrics=["accuracy"])

model_B.summary()
```

```python
import tensorflow as tf
from tensorflow import keras

print("=== 1. еҮҶеӨҮиөӣйҒ“пјҡеҠ иҪҪе№¶жё…жҙ—ж•°жҚ® ===")
fashion_mnist = keras.datasets.fashion_mnist
(X_train_full, y_train_full), (X_test, y_test) = fashion_mnist.load_data()

# жһҒе…¶йҮҚиҰҒпјҡжҠҠ 0-255 зҡ„еғҸзҙ еҖјзј©ж”ҫеҲ° 0-1 д№Ӣй—ҙпјҢзҘһз»ҸзҪ‘з»ңжңҖе–ңж¬ўиҝҷз§ҚеӨ§е°Ҹзҡ„ж•°еӯ—
X_train_full = X_train_full / 255.0
X_test = X_test / 255.0

# еҲҮеҲҶеҮәйӘҢиҜҒйӣҶ (Validation Set)пјҢз”ЁжқҘз»ҷвҖңжҷәиғҪйӣ·иҫҫвҖқеҪ“еҸӮиҖғ
X_valid, X_train = X_train_full[:5000], X_train_full[5000:]
y_valid, y_train = y_train_full[:5000], y_train_full[5000:]


print("\n=== 2. з»„иЈ…и¶…и·‘еә•зӣҳпјҡе®ҢзҫҺжһ¶жһ„жҗӯе»ә ===")
model = keras.models.Sequential([
    keras.Input(shape=(28, 28)), # Keras 3 ж ҮеҮҶе…ҘеҸЈ
    keras.layers.Flatten(),
    
    # з¬¬дёҖеұӮйҡҗи—ҸеұӮпјҡHe еҲқе§ӢеҢ– + жү№еҪ’дёҖеҢ– (BN) + ELU жҝҖжҙ»еҮҪж•°
    keras.layers.Dense(300, kernel_initializer="he_normal"),
    keras.layers.BatchNormalization(), # BN еұӮж”ҫеңЁ Dense е’Ң Activation д№Ӣй—ҙж•ҲжһңжһҒдҪі
    keras.layers.Activation("elu"),
    
    # з¬¬дәҢеұӮйҡҗи—ҸеұӮпјҡеҗҢж ·зҡ„ж»Ўй…ҚзҘһиЈ…
    keras.layers.Dense(100, kernel_initializer="he_normal"),
    keras.layers.BatchNormalization(),
    keras.layers.Activation("elu"),
    keras.layers.Dropout(rate=0.2),
    # иҫ“еҮәеұӮпјҡ10 з§ҚиЎЈжңҚпјҢжүҖд»ҘжҳҜ 10 дёӘзҘһз»Ҹе…ғпјҢSoftmax иҫ“еҮәжҰӮзҺҮ
    keras.layers.Dense(10, activation="softmax")
])


print("\n=== 3. иЈ…иҪҪ V8 еј•ж“ҺпјҡдјҳеҢ–еҷЁдёҺзј–иҜ‘ ===")
# жҲ‘д»¬дёҚз”Ёжҷ®йҖҡзҡ„ AdamпјҢзӣҙжҺҘдёҠеёҰвҖңжҸҗеүҚеҲ№иҪҰйӣ·иҫҫ(Nesterov)вҖқзҡ„еҚҮзә§зүҲ NadamпјҒ
optimizer = keras.optimizers.Nadam(learning_rate=0.001)

model.compile(loss="sparse_categorical_crossentropy",
              optimizer=optimizer,
              metrics=["accuracy"])


print("\n=== 4. й…ҚзҪ®иҮӘеҠЁй©ҫй©¶д»Әпјҡеӣһи°ғеҮҪж•° (Callbacks) ===")
# жҷәиғҪйӣ·иҫҫ 1пјҡReduceLROnPlateau (еҠЁжҖҒеӯҰд№ зҺҮ)
# еҸӘиҰҒйӘҢиҜҒйӣҶиҜҜе·®иҝһз»ӯ 4 иҪ®жІЎйҷҚпјҢеӯҰд№ зҺҮзӣҙжҺҘз ҚеҚҠ (д№ҳд»Ҙ 0.5)
lr_scheduler = keras.callbacks.ReduceLROnPlateau(
    monitor="val_loss", factor=0.5, patience=4, verbose=1
)

# жҷәиғҪйӣ·иҫҫ 2пјҡEarlyStopping (ж—©еҒңжі•пјҢдҝқе‘ҪзҘһжҠҖ)
# еҰӮжһңиҝһз»ӯ 10 иҪ®йғҪжІЎиҝӣжӯҘпјҢзӣҙжҺҘејәиЎҢжӢ”з”өжәҗеҒңжӯўи®ӯз»ғпјҢе№¶жҠҠжңҖеҘҪзҡ„жқғйҮҚиҝҳз»ҷдҪ пјҒ
early_stopping = keras.callbacks.EarlyStopping(
    monitor="val_loss", patience=10, restore_best_weights=True
)


print("\n=== 5. иё©дёӢжІ№й—ЁпјҢжһҒйҖҹзӢӮйЈҷпјҒ ===")
# жҠҠж•°жҚ®гҖҒйӘҢиҜҒйӣҶгҖҒд»ҘеҸҠжҲ‘д»¬ејәеӨ§зҡ„вҖңжҷәиғҪйӣ·иҫҫвҖқз»ҹз»ҹжҢӮиҪҪдёҠеҺ»
history = model.fit(
    X_train, y_train, 
    epochs=200, # дҪ ж•ўи®ҫ 50 иҪ®пјҢжҳҜеӣ дёәжңү EarlyStopping дҝқжҠӨпјҢз»қдёҚдјҡиҝҮжӢҹеҗҲеҲ°жӯ»
    validation_data=(X_valid, y_valid),
    callbacks=[lr_scheduler, early_stopping]
)

print("\n=== и®ӯз»ғз»“жқҹпјҒеҸҜд»ҘеңЁжөӢиҜ•йӣҶдёҠйӘҢж”¶жҲҗжһңдәҶ ===")
test_loss, test_acc = model.evaluate(X_test, y_test)
print(f"жңҖз»ҲжөӢиҜ•йӣҶеҮҶзЎ®зҺҮ: {test_acc:.4f}")
```

```python
model = keras.models.Sequential()

model.add(keras.layers.Input(shape=[32, 32, 3]))
model.add(keras.layers.Flatten())

for i in range(20):
    model.add(keras.layers.Dense(100,kernel_initializer="he_normal"))
    model.add(keras.layers.BatchNormalization())
    model.add(keras.layers.Activation("elu"))

model.add(keras.layers.Dense(10, activation="softmax"))

optimizer_nadam = keras.optimizers.Nadam(learning_rate=5e-5)

model.compile(
    loss="sparse_categorical_crossentropy",
    optimizer=optimizer_nadam,
    metrics=["accuracy"]
)

log_dir = os.path.join("logs", "fit")
tensorboard_cb = keras.callbacks.TensorBoard(
    log_dir=log_dir, 
    histogram_freq=0  # жҜҸдёҖиҪ®йғҪз»ҹи®ЎдёҖдёӢжқғйҮҚзҡ„зӣҙж–№еӣҫпјҲеҸҜйҖүпјҢжғізңӢиҜҰз»Ҷзҡ„е°ұи®ҫдёә1пјү
)

callbacks=[keras.callbacks.EarlyStopping(patience=5, restore_best_weights=True),
           keras.callbacks.ModelCheckpoint("mnist_best.keras", save_best_only=True),
          tensorboard_cb]

history = model.fit(X_train,y_train,epochs=100,
                    validation_data=(X_valid,y_valid),
                   callbacks=callbacks)
```

```python
import tensorflow as tf
from tensorflow import keras

model = keras.models.Sequential([
    # рҹҢҹ з¬¬дёҖеұӮпјҡдёҚеҶҚжҳҜ FlattenпјҒзӣҙжҺҘз”Ё 64 дёӘ 3x3 зҡ„ж”ҫеӨ§й•ңеҺ»жү«жҸҸ 32x32 зҡ„еҪ©иүІеӣҫзүҮпјҒ
    # еЎ«е…… padding="same" ж„Ҹе‘ізқҖжү«жҸҸеҗҺеӣҫзүҮеӨ§е°ҸдёҚеҸҳгҖӮ
    keras.layers.Conv2D(64, kernel_size=3, activation="relu", padding="same", input_shape=[32, 32, 3]),
    
    # рҹ—ңпёҸ еҺӢзј©жңәпјҡй•ҝе®Ҫеҗ„з ҚдёҖеҚҠпјҲеҸҳжҲҗ 16x16пјү
    keras.layers.MaxPooling2D(pool_size=2),
    
    # рҹҢҹ з¬¬дәҢеұӮпјҡз”Ё 128 дёӘж”ҫеӨ§й•ңз»§з»ӯжү«жҸҸиҝҷзј©е°ҸеҗҺзҡ„ 16x16 еӣҫзүҮ
    keras.layers.Conv2D(128, kernel_size=3, activation="relu", padding="same"),
    
    # рҹ—ңпёҸ еҺӢзј©жңәпјҡй•ҝе®ҪеҶҚз ҚдёҖеҚҠпјҲеҸҳжҲҗ 8x8пјү
    keras.layers.MaxPooling2D(pool_size=2),
    
    # рҹ§ұ жңҖеҗҺж”¶е°ҫпјҡз»ҸиҝҮдәҶеүҚйқўзҡ„жө“зј©жҸҗзӮјпјҢжҲ‘д»¬зҺ°еңЁеҸҜд»Ҙж”ҫеҝғең°жҠҠе®ғжӢҚжүҒдәҶ
    keras.layers.Flatten(),
    
    # з”Ёжҷ®йҖҡзҡ„ Dense еұӮеҒҡжңҖеҗҺзҡ„ 10 еҲҶзұ»иҫ“еҮә
    keras.layers.Dense(64, activation="relu"),
    keras.layers.Dense(10, activation="softmax")
])
```

```python
import tensorflow as tf
from tensorflow import keras

print("=== 1. еҸ¬е”Ө CIFAR-10 ж•°жҚ®йӣҶ ===")
cifar10 = keras.datasets.cifar10
(X_train_full, y_train_full), (X_test, y_test) = cifar10.load_data()

# жһҒе…¶е…ій”®пјҡе°Ҷ 0-255 зҡ„еғҸзҙ еҖјеҺӢзј©еҲ° 0-1 д№Ӣй—ҙ
X_train_full = X_train_full / 255.0
X_test = X_test / 255.0

# еүІиӮүпјҡеҲҮеҲҶеҮә 5000 еј дҪңдёәйӘҢиҜҒйӣҶ
X_valid, X_train = X_train_full[:5000], X_train_full[5000:]
y_valid, y_train = y_train_full[:5000], y_train_full[5000:]


print("\n=== 2. жҗӯе»ә CNN ж ёеҝғжһ¶жһ„ ===")
model = keras.models.Sequential([
    # рҹҢҹ з¬¬дёҖеұӮпјҡ64 дёӘ 3x3 зҡ„ж”ҫеӨ§й•ңпјҢдҝқз•ҷиҫ№зјҳеЎ«е…… (same)
    keras.layers.Conv2D(64, kernel_size=3, activation="relu", padding="same", input_shape=[32, 32, 3]),
    # рҹ—ңпёҸ еҺӢзј©жңәпјҡй•ҝе®Ҫз ҚеҚҠ (32x32 еҸҳжҲҗ 16x16)
    keras.layers.MaxPooling2D(pool_size=2),
    
    # рҹҢҹ з¬¬дәҢеұӮпјҡ128 дёӘж”ҫеӨ§й•ңз»§з»ӯжҸҗзӮјй«ҳзә§зү№еҫҒ
    keras.layers.Conv2D(128, kernel_size=3, activation="relu", padding="same"),
    # рҹ—ңпёҸ еҺӢзј©жңәпјҡй•ҝе®ҪеҶҚз ҚеҚҠ (16x16 еҸҳжҲҗ 8x8)
    keras.layers.MaxPooling2D(pool_size=2),
    
    # рҹ§ұ ж”¶е°ҫпјҡжӯӨж—¶еӣҫзүҮе·Із»Ҹиў«жө“зј©жҲҗдәҶ 8x8 зҡ„жһҒе“Ғзү№еҫҒеӣҫпјҢеҸҜд»Ҙж”ҫеҝғжӢҚжүҒдәҶ
    keras.layers.Flatten(),
    keras.layers.Dense(64, activation="relu"),
    keras.layers.Dense(10, activation="softmax")
])

# жү“еҚ°дёҖдёӢеӨ§жҘјз»“жһ„пјҢдҪ дјҡеҸ‘зҺ°еҸӮж•°йҮҸжҜ”д№ӢеүҚе°‘пјҢдҪҶз»“жһ„жһҒе…¶зІҫжӮҚпјҒ
model.summary()


print("\n=== 3. жҢӮиҪҪеј•ж“ҺдёҺйӣ·иҫҫ ===")
# еӣ дёә CNN жһҒе…¶зЁіе®ҡпјҢжҲ‘д»¬з”ҡиҮідёҚйңҖиҰҒжҗһеҮ дёҮеҲҶд№ӢдёҖзҡ„еҫ®е°ҸеӯҰд№ зҺҮпјҢзӣҙжҺҘз”Ёй»ҳи®Өзҡ„ Nadam е°ұиЎҢпјҒ
model.compile(loss="sparse_categorical_crossentropy",
              optimizer="nadam",
              metrics=["accuracy"])

# дҝқе‘Ҫйӣ·иҫҫ (иҖҗеҝғеҖји®ҫдёә 5пјҢжӢҝеҲ°жңҖеҘҪжқғйҮҚ)
early_stopping_cb = keras.callbacks.EarlyStopping(patience=5, restore_best_weights=True)


print("\n=== 4. рҹҡҖ зӮ№зҒ«иө·йЈһпјҒдҪ“йӘҢйҷҚз»ҙжү“еҮ» ===")
# е…Ҳи®ҫе®ҡи·‘ 15 иҪ®пјҢдҪ зӣҜзқҖжҜҸдёҖиҪ®зҡ„еҮҶзЎ®зҺҮзңӢпјҒ
history = model.fit(X_train, y_train, epochs=15,
                    validation_data=(X_valid, y_valid),
                    callbacks=[early_stopping_cb])


print("\n=== 5. з»ҲжһҒйӘҢж”¶ ===")
test_loss, test_acc = model.evaluate(X_test, y_test)
print(f"рҹҺү CNN йҰ–ж¬ЎеҮәжҲҳпјҢCIFAR-10 жөӢиҜ•йӣҶжңҖз»ҲеҮҶзЎ®зҺҮ: {test_acc:.4f}")
```
