```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import signal

# ==========================================
# 1. 生成模拟的 CMJ 多肌肉数据
# ==========================================
np.random.seed(42)
fs = 1000
duration = 1.5  # 截取动作核心部分 1.5秒
t = np.linspace(0, duration, int(fs * duration))

# 定义肌肉爆发模式 (蹬伸阶段在 0.6s - 0.9s)
def create_burst(t, center, width):
    return np.exp(-((t - center)**2) / (2 * width**2))

# 模拟共享的神经驱动 (Common Drive) - 这是产生相干性的源头
# 假设有一个 20Hz (Beta波) 的神经振荡信号控制着所有伸肌
common_drive_beta = 0.5 * np.sin(2 * np.pi * 20 * t) 

# 生成各肌肉信号 (基线噪音 + 爆发包络 + 共同驱动)
muscles = ['RF', 'VM', 'BF', 'GM', 'TA', 'GMed']
data = {}

for m in muscles:
    # 基础包络 (模拟用力)
    if m in ['RF', 'VM', 'GM', 'GMed']: # 伸肌/原动肌群
        env = create_burst(t, 0.75, 0.1) 
        # 伸肌群混入较强的共同驱动 (会导致高相干性)
        drive = common_drive_beta * env 
    else: # 拮抗肌 (BF, TA)
        env = create_burst(t, 0.75, 0.1) * 0.3 # 激活低
        drive = np.random.normal(0, 0.1, len(t)) * env # 没什么共同驱动
    
    # 原始信号 = (包络 + 驱动) * 高斯白噪 (载波)
    # 注意: 真实的相干性通常隐藏在解调后的包络或特定的频段里
    noise = np.random.normal(0, 1, len(t))
    raw_sig = (env + 0.5 * drive) * noise 
    data[m] = raw_sig

df = pd.DataFrame(data)

# ==========================================
# 2. 核心处理函数
# ==========================================

def compute_connectivity_matrices(df, fs):
    """
    计算两种矩阵:
    1. 时域相关矩阵 (Envelope Correlation) - 反映协同发力
    2. 频域相干矩阵 (Beta-band Coherence) - 反映神经驱动同步
    """
    muscles = df.columns
    n = len(muscles)
    
    # 初始化矩阵
    corr_matrix = np.zeros((n, n))
    coh_matrix = np.zeros((n, n))
    
    # --- 预处理: 获取线性包络 (用于时域相关) ---
    envelopes = pd.DataFrame()
    for m in muscles:
        # 带通滤波 -> 整流 -> 低通滤波(得到平滑包络)
        b, a = signal.butter(4, [20/500, 450/500], btype='band')
        filt = signal.filtfilt(b, a, df[m])
        rect = np.abs(filt)
        # 10Hz 低通滤波取包络
        b_low, a_low = signal.butter(4, 10/500, btype='low')
        env = signal.filtfilt(b_low, a_low, rect)
        envelopes[m] = env

    # --- 预处理: 整流信号 (用于频域相干) ---
    # 文献争议点: 多数研究建议对 EMG 整流后再做相干性，能增强 Beta 频段检测
    rectified_data = df.apply(lambda x: np.abs(signal.filtfilt(b, a, x)))

    # --- 双重循环计算两两配对 ---
    for i in range(n):
        for j in range(n):
            m1 = muscles[i]
            m2 = muscles[j]
            
            if i == j:
                corr_matrix[i, j] = 1.0
                coh_matrix[i, j]  = 1.0 # 自身相干为1 (或设为NaN以便绘图)
                continue
            
            # 1. 计算时域相关 (Pearson)
            r = np.corrcoef(envelopes[m1], envelopes[m2])[0, 1]
            corr_matrix[i, j] = r
            
            # 2. 计算频域相干 (Magnitude Squared Coherence)
            # 使用 Welch 方法
            f, Cxy = signal.coherence(rectified_data[m1], rectified_data[m2], fs, nperseg=256)
            
            # 提取 Beta 频段 (15-35 Hz) 的平均相干值
            beta_idx = (f >= 15) & (f <= 35)
            mean_beta_coh = np.mean(Cxy[beta_idx])
            coh_matrix[i, j] = mean_beta_coh

    return corr_matrix, coh_matrix, muscles

# 运行计算
corr_mat, coh_mat, names = compute_connectivity_matrices(df, fs)

# ==========================================
# 3. 可视化结果 (绘制热力图)
# ==========================================
fig, ax = plt.subplots(1, 2, figsize=(16, 7))

# 图 A: 时域协同 (Envelope Correlation)
sns.heatmap(corr_mat, annot=True, fmt=".2f", xticklabels=names, yticklabels=names, 
            cmap="Reds", vmin=0, vmax=1, ax=ax[0], square=True)
ax[0].set_title("A. Time-Domain Synergy\n(Envelope Correlation)", fontsize=14, fontweight='bold')

# 图 B: 频域相干 (Beta-band Coherence)
sns.heatmap(coh_mat, annot=True, fmt=".2f", xticklabels=names, yticklabels=names, 
            cmap="Blues", vmin=0, vmax=0.5, ax=ax[1], square=True) # 相干值通常较低，vmax设0.5即可
ax[1].set_title("B. Neural Synchrony\n(Beta-band 15-35Hz Coherence)", fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
```
