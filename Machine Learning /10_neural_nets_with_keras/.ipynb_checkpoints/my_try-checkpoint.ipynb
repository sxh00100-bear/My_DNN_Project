```python
import tensorflow as tf
from tensorflow import keras

# 1. 准备数据 (还是那堆衣服)
fashion_mnist = keras.datasets.fashion_mnist
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()
# 归一化：把像素值从 0-255 压缩到 0-1 (为了让公式算得更稳)
X_train, X_test = X_train / 255.0, X_test / 255.0

# 2. 搭建模型 (此处就在应用公式 10-2)
model = keras.models.Sequential([
    # 第一步：把 28x28 的二维图拉平成 784 的一维向量 (这是输入 X)
    keras.layers.Flatten(input_shape=[28, 28]),
    
    # --- 隐藏层 1 ---
    # 300个神经元。这里发生了：ReLU(X * W1 + b1)
    # W1 的形状是 (784, 300)，b1 的形状是 (300,)
    keras.layers.Dense(300, activation="relu"),
    
    # --- 隐藏层 2 ---
    # 100个神经元。这里发生了：ReLU(输出1 * W2 + b2)
    # W2 的形状是 (300, 100)，b2 的形状是 (100,)
    keras.layers.Dense(100, activation="relu"),
    
    # --- 输出层 ---
    # 10个神经元。这里发生了：Softmax(输出2 * W3 + b3)
    keras.layers.Dense(10, activation="softmax")
])

# 3. 看看刚才那个公式到底生成了多少参数 (W和b)
print("🔍 模型的参数统计：")
model.summary()

# 4. 编译并训练 (让 SGD 去调整 W 和 b)
print("\n🚀 开始在 M3 上训练...")
model.compile(loss="sparse_categorical_crossentropy",
              optimizer="sgd",
              metrics=["accuracy"])

# 跑 5 轮，见证准确率上涨
history = model.fit(X_train, y_train, epochs=5)
```

```python
# 1. 让模型预测测试集的前 3 张图
X_new = X_test[:10]
y_proba = model.predict(X_new)

# 2. 打印预测结果 (概率)
print("\n🔍 模型的心理活动 (概率分布):")
print(y_proba.round(2)) # 保留两位小数

# 3. 翻译成人类能懂的类别
import numpy as np
class_names = ["T-shirt/top", "Trouser", "Pullover", "Dress", "Coat",
               "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"]

y_pred = np.argmax(y_proba, axis=1) # 找概率最大的那个位置
print("\n🏆 最终预测结果:")
print(np.array(class_names)[y_pred])

# 4. 看看标准答案对不对
print("\n✅ 真实标签 (标准答案):")
print(np.array(class_names)[y_test[:10]])
```

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt

# 1. 加载 MNIST (手写数字 0-9)
(X_train_full, y_train_full), (X_test, y_test) = keras.datasets.mnist.load_data()

# 2. 归一化 (0-255 -> 0-1)
X_train, X_test = X_train_full / 255.0, X_test / 255.0

# 3. 随便挑一张测试图出来 (比如第 15 张图，它真实数字是 5)
sample_img = X_test[15:16]  # 保持 (1, 28, 28) 的形状
print(f"这张图的真实标签是: {y_test[15]}")
plt.imshow(sample_img[0], cmap="binary")
plt.axis('off')
plt.show()
```

```python
# --- 数据准备：制作“二分类”标签 ---
# 如果是 5 就是 1，不是 5 就是 0
y_train_5 = (y_train_full == 5)
y_test_5 = (y_test == 5)

# --- 搭建模型 ---
model_binary = keras.models.Sequential([
    keras.layers.Flatten(input_shape=[28, 28]),
    keras.layers.Dense(100, activation="relu"),
    
    # 【关键点】：1个神经元，Sigmoid 激活
    # 输出范围 0~1，代表“是5的可能性”
    keras.layers.Dense(1, activation="sigmoid") 
])

# --- 编译与训练 (简单跑 2 轮演示) ---
model_binary.compile(loss="binary_crossentropy", optimizer="sgd", metrics=["accuracy"])
model_binary.fit(X_train, y_train_5, epochs=2, verbose=0) # verbose=0 不打印刷屏日志

# --- 预测演示 ---
# 这里的 output 直接就是概率
prob = model_binary.predict(sample_img)

print(f"\n【二分类预测结果】")
print(f"预测概率值: {prob[0][0]:.4f}")
print(f"最终判断: {'是 5' if prob[0][0] > 0.5 else '不是 5'}")
```

```python
# --- 搭建模型 ---
model_multiclass = keras.models.Sequential([
    keras.layers.Flatten(input_shape=[28, 28]),
    keras.layers.Dense(100, activation="relu"),
    
    # 【关键点】：10个神经元，Softmax 激活
    # 输出 10 个概率值，加起来等于 1
    keras.layers.Dense(10, activation="softmax")
])

# --- 编译与训练 ---
# 注意 loss 变成了 sparse_categorical_crossentropy
model_multiclass.compile(loss="sparse_categorical_crossentropy", optimizer="sgd", metrics=["accuracy"])
model_multiclass.fit(X_train, y_train_full, epochs=2, verbose=0)

# --- 预测演示 ---
predictions = model_multiclass.predict(sample_img)

print(f"\n【多分类预测结果】")

# 需求 A: 有多大的概率是数字几？(0-9)
# 我们把 10 个概率都打印出来
print("1. 概率分布 (0-9):")
with np.printoptions(precision=4, suppress=True): # 格式化一下打印效果，不显示科学计数法
    print(predictions[0])

# 需求 B: 到底是哪个数字？
# 使用 argmax 找到概率最大的那个位置的索引
predicted_class = np.argmax(predictions, axis=1)
print(f"\n2. 最终判定类别: {predicted_class[0]}")
print(f"   (置信度: {np.max(predictions):.2%})")
```

```python
import pandas as pd
import numpy as np
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# 1. 加载原始数据
housing = fetch_california_housing()
X_all = housing.data
y_all = housing.target

# 2. 切分数据集 (训练集 : 验证集 : 测试集)
# 先切出 测试集
X_train_full, X_test, y_train_full, y_test = train_test_split(
    X_all, y_all, random_state=42
)
# 再把剩下的切出 验证集
X_train, X_valid, y_train, y_valid = train_test_split(
    X_train_full, y_train_full, random_state=42
)

# 3. 数据标准化 (非常重要！如果不做，Loss 可能会爆炸)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_valid = scaler.transform(X_valid)
X_test = scaler.transform(X_test)

# ==========================================
# 4. 【关键步骤】把数据切成两份 (Wide 和 Deep)
# ==========================================
# 假设原始数据有 8 列。
# Wide 路走前 5 列 (索引 0~4)
# Deep 路走后 6 列 (索引 2~7) --> 这样有些特征是重叠的，符合 Wide&Deep 的设计
# ==========================================

# --- 训练集 ---
X_train_wide = X_train[:, :5]   # 取前5个特征
X_train_deep = X_train[:, 2:]   # 取从第2个开始到最后的特征 (共6个)

# --- 验证集 ---
X_valid_wide = X_valid[:, :5]
X_valid_deep = X_valid[:, 2:]

# --- 测试集 ---
X_test_wide = X_test[:, :5]
X_test_deep = X_test[:, 2:]

# --- 新数据 (用于最后的 predict 演示) ---
X_new_wide = X_test_wide[:3]
X_new_deep = X_test_deep[:3]

# 5. 打印一下形状，确保和你的模型对应得上
print(f"Wide 输入形状: {X_train_wide.shape}") # 应该是 (..., 5)
print(f"Deep 输入形状: {X_train_deep.shape}") # 应该是 (..., 6)
print("✅ 数据准备完毕！现在可以运行 model.fit 了")
```

```python
import tensorflow as tf
```

```python
class WideAndDeepModel(tf.keras.Model):
    def __init__(self, units=30, activation="relu", **kwargs):
        super().__init__(**kwargs)  # needed to support naming the model
        self.norm_layer_wide = tf.keras.layers.Normalization()
        self.norm_layer_deep = tf.keras.layers.Normalization()
        self.hidden1 = tf.keras.layers.Dense(units, activation=activation)
        self.hidden2 = tf.keras.layers.Dense(units, activation=activation)
        self.main_output = tf.keras.layers.Dense(1)
        self.aux_output = tf.keras.layers.Dense(1)
        
    def call(self, inputs):
        input_wide, input_deep = inputs
        norm_wide = self.norm_layer_wide(input_wide)
        norm_deep = self.norm_layer_deep(input_deep)
        hidden1 = self.hidden1(norm_deep)
        hidden2 = self.hidden2(hidden1)
        concat = tf.keras.layers.concatenate([norm_wide, hidden2])
        output = self.main_output(concat)
        aux_output = self.aux_output(hidden2)
        return output, aux_output

tf.random.set_seed(42)  # extra code – just for reproducibility
model = WideAndDeepModel(30, activation="relu", name="my_cool_model")
```

```python
callbacks = [
    tf.keras.callbacks.EarlyStopping(
        monitor = 'val_loss',patience = 5,restore_best_weights=True),
    tf.keras.callbacks.ModelCheckpoint(
        "best_weight.weights.h5",monitor='val_loss',save_best_only=True,save_weights_only=True),
    tf.keras.callbacks.TensorBoard(
        log_dir = 'logs')
]
model.norm_layer_wide.adapt(X_train_wide)
model.norm_layer_deep.adapt(X_train_deep)
optimizer = tf.keras.optimizers.Adam(learning_rate=5e-4)
model.compile(loss=["mse", "mse"], loss_weights=[0.9, 0.1], optimizer=optimizer,
              metrics=["RootMeanSquaredError", "RootMeanSquaredError"])

history = model.fit(
    (X_train_wide, X_train_deep), (y_train, y_train), epochs=100,
    validation_data=((X_valid_wide, X_valid_deep), (y_valid, y_valid)),
    callbacks=callbacks, verbose=1)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# 1. 把训练记录转换成表格
pd.DataFrame(history.history).plot(figsize=(10, 6))

# 2. 设置一下样式，好看一点
plt.grid(True) # 加网格
plt.gca().set_ylim(0, 1) # 设置Y轴范围 (因为你的Loss大概在0.3左右，设为0-1看得很清楚)
plt.title("Training History") # 标题
plt.xlabel("Epochs") # X轴是轮数
plt.ylabel("Loss / Error") # Y轴是分数

# 3. 展示
plt.show()
```
