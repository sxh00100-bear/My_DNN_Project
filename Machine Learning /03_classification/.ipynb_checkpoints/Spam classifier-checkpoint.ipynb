```python
import os
import tarfile
import urllib.request
from pathlib import Path

def fetch_spam_data():
    # 1. 设置下载根目录和文件名 (这里我改成了你刚才问的 2002 版本)
    spam_root = "https://spamassassin.apache.org/old/publiccorpus/"
    ham_url = spam_root + "20021010_easy_ham.tar.bz2"
    spam_url = spam_root + "20021010_spam.tar.bz2"

    # 2. 设置本地存储路径: datasets/spam
    spam_path = Path() / "datasets" / "spam"
    spam_path.mkdir(parents=True, exist_ok=True)

    # 3. 循环下载并解压
    # 列表格式: (解压后的文件夹名, 下载保存的文件名, 下载链接)
    for dir_name, tar_name, url in (("easy_ham", "ham", ham_url),
                                    ("spam", "spam", spam_url)):
        
        # 检查是否已经存在，避免重复下载
        if not (spam_path / dir_name).is_dir():
            path = (spam_path / tar_name).with_suffix(".tar.bz2")
            print(f"Downloading {tar_name}...")
            urllib.request.urlretrieve(url, path)
            
            # 解压
            with tarfile.open(path) as tar_bz2_file:
                tar_bz2_file.extractall(path=spam_path)
    
    # 返回解压后的文件夹路径列表
    return [spam_path / dir_name for dir_name in ("easy_ham", "spam")]

# 运行函数
ham_dir, spam_dir = fetch_spam_data()
print("Success! Data is ready.")
print("Ham directory:", ham_dir)
print("Spam directory:", spam_dir)
```

```python
ham_filenames = [f for f in sorted(ham_dir.iterdir()) if len(f.name) > 20]
spam_filenames = [f for f in sorted(spam_dir.iterdir()) if len(f.name) > 20]
#“请把文件夹里所有的文件都拿出来，按顺序排好。但是，名字太短的文件（比如那个讨厌的 cmds）我不要，我只要那些长长名字（长于20）的邮件文件。”
```

```python
import email
import email.policy

def load_email(filepath):
    with open(filepath, "rb") as f:
        return email.parser.BytesParser(policy=email.policy.default).parse(f)
```

```python
ham_emails = [load_email(filepath) for filepath in ham_filenames]
spam_emails = [load_email(filepath) for filepath in spam_filenames]
```
