```python
import numpy as np

# ==========================================
# 模拟真实肌电信号生成器
# ==========================================
def get_mock_emg_data():
    fs = 1000.0           # 采样率 1000Hz
    duration = 2.0        # 时长 2秒
    n_points = int(fs * duration)
    t = np.linspace(0, duration, n_points)
    
    # 1. 基础白噪音 (模拟环境电磁干扰和人体背景噪音)
    # 幅度很小，大约在 +/- 0.02mV
    noise = np.random.normal(0, 0.02, n_points)
    
    # 2. 模拟射门时的肌肉爆发 (Burst)
    # 我们用一个高斯窗函数(Gaussian Window)来模拟肌肉从放松到紧张再到放松的过程
    # 假设射门发生在 1.0秒左右，持续约 0.3秒
    burst_center = 1.0
    burst_width = 0.1
    burst_envelope = np.exp(-0.5 * ((t - burst_center) / burst_width)**2)
    
    # 生成爆发时的肌电信号 (高频杂波 * 爆发包络)
    # 假设最大发力电压能达到 2.5mV
    muscle_activity = np.random.normal(0, 1, n_points) * burst_envelope * 2.5
    
    # 3. 叠加信号
    raw_signal = noise + muscle_activity
    
    return t, raw_signal, fs

# ==========================================
# 使用方法
# ==========================================
# 调用这个函数，你就得到了 t (时间轴), raw_data (原始信号), fs (采样率)
t, raw_data, fs = get_mock_emg_data()

# 打印前10个数据看看长什么样
print("原始信号样本 (前10个点):")
print(raw_data[:10])
```

```python
import numpy as np

import matplotlib.pyplot as plt

# 1. 设置字体为 Mac 通用中文宽字体
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS'] 
# 如果上面这个不行，尝试换成 ['Heiti TC']

# 2. 解决负号显示问题
plt.rcParams['axes.unicode_minus'] = False
from scipy.signal import butter, filtfilt

# --- 1. 这里放入刚才的数据生成器 ---
def get_mock_emg_data():
    fs = 1000.0
    duration = 2.0
    n_points = int(fs * duration)
    t = np.linspace(0, duration, n_points)
    noise = np.random.normal(0, 0.02, n_points) # 底噪
    # 模拟射门爆发：在1秒处，最大振幅2.5mV
    burst_envelope = np.exp(-0.5 * ((t - 1.0) / 0.1)**2)
    muscle_activity = np.random.normal(0, 1, n_points) * burst_envelope * 2.5
    return t, noise + muscle_activity, fs

# --- 2. 这里是之前给你的核心处理函数 ---
def process_emg_signal(raw_data, fs):
    # A. 去直流
    signal_no_dc = raw_data - np.mean(raw_data)
    # B. 带通滤波 (20-450Hz)
    nyq = 0.5 * fs
    b, a = butter(4, [20.0/nyq, 450.0/nyq], btype='bandpass')
    signal_bandpass = filtfilt(b, a, signal_no_dc)
    # C. 整流
    signal_rectified = np.abs(signal_bandpass)
    # D. 线性包络 (6Hz低通)
    b_env, a_env = butter(4, 6.0/nyq, btype='lowpass')
    signal_envelope = filtfilt(b_env, a_env, signal_rectified)
    return signal_envelope

# --- 3. 主程序 ---
if __name__ == "__main__":
    # 获取模拟数据
    t, raw_data, fs = get_mock_emg_data()
    
    # 处理数据
    envelope = process_emg_signal(raw_data, fs)
    
    # 画图
    plt.figure(figsize=(12, 6))
    
    # 上图：原始乱糟糟的信号
    plt.subplot(2, 1, 1)
    plt.plot(t, raw_data, color='#555555', linewidth=0.8)
    plt.title(f"原始肌电信号 (Raw EMG) - 模拟射门瞬间")
    plt.ylabel("电压 (mV)")
    plt.grid(alpha=0.3)
    
    # 下图：处理后清晰的包络线
    plt.subplot(2, 1, 2)
    plt.plot(t, envelope, color='#D32F2F', linewidth=2)
    plt.fill_between(t, envelope, color='#D32F2F', alpha=0.2) # 填充颜色更好看
    plt.title("处理后的线性包络 (Linear Envelope)")
    plt.ylabel("激活程度 (mV)")
    plt.grid(alpha=0.3)
    
    plt.tight_layout()
    plt.show()
```
